### Concating the DF:

```python
import pandas as pd
```

```python
df1=pd.read_csv("project_adv.csv")
df2=pd.read_csv("project_rom.csv")
df3=pd.read_csv("project_scifi.csv")
df4=pd.read_csv("project_music.csv")
df5=pd.read_csv("project_his.csv")
```
